#### Explore the artist
ideas:
* top occurring names
* least occurring names
* collaboration between artists
* longest name

```python
px.histogram(df, x='Artist', title='')
```

```python
# your code here
df['Artist'].value_counts().head(10)
```

```python
# your code here
top_atists = df['Artist'].value_counts()
print(type(top_atists))
top_atists[top_atists == 1]
```

```python
px.bar(top_atists, x=top_atists.index, y=top_atists.values, title='')
```

```python
df_top_artist = pd.DataFrame(top_atists).reset_index()
df_top_artist[df_top_artist['count'] == 1]
```

```python
px.bar(df_top_artist.head(30), x='Artist', y='count', title='')
```

Get the top artist's unique albums

```python
max_count = df_top_artist['count'].max()
df_top_artist[df_top_artist['count'] == max_count]
```

```python
top_artist = df_top_artist[df_top_artist['count'] == max_count]['Artist'].tolist()
top_artist
```

Which album has the most tracks?

```python
df[df['Artist'].isin(top_artist)][['Artist', 'Album Name']].drop_duplicates().sort_values('Artist')
```

* oldest album
* newest album

```python
df['Release Date'] = pd.to_datetime(df['Release Date'])
```

```python
min_year = df['Release Date'].dt.year.min()
df[df['Release Date'].dt.year == min_year]
```

plot the distribution of albums over time

```python
px.histogram(df, x='Release Date', title='')
```

```python
# month with most releases
df['release_month'] = df['Release Date'].dt.month
px.histogram(df, x='release_month', title='')
```

```python
df['release_year'] = df['Release Date'].dt.year
px.histogram(df, x='release_year', title='')
```

```python
px.histogram(df[df['release_year'] != 2024], x='release_month', title='')
```

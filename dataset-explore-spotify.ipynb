```python
import zipfile
import chardet

# Open the ZIP file
with zipfile.ZipFile('Spotify Top Songs 2024.zip', 'r') as zip_ref:
    # Extract the file name
    file_name = zip_ref.namelist()[0]  # Assuming there's only one file in the zip
    with zip_ref.open(file_name) as f:
        # Read the file contents in binary mode
        file_data = f.read()
        # Detect the encoding
        result = chardet.detect(file_data)

result
```
